```python
#to select the specific columns for analysis
#To analyse the variation of "Depth vs Temp" & "Depth Vs Salinity" of the data "bottle.csv" 
sel=df[['Depthm','T_degC', 'Salnty']]
sel.head(29) #command to select the 29 sample data points for analysizing the variations
```

```python
#Command to rename the columns 
sel.columns=['Depth_metre', 'Temperature_C', 'Salinity']
sel.head(29)
```

```python
sel1=sel.iloc[:29] #Selecting pandas data-used for location based indexing
```

```python
sel1.head(2)# command to display the first two rows
```

```python
#to check the info of the DataFrame
sel1.info() 
# therefore 1 NAN values is present in the Salinity Column as 28 non null values are present 
#while in all other column 29 non null values, so we have to replace one null value(NAN)
```

```python
#another command to check the presence of null values
sel1.isna()
```

```python
#It is clear from the above that the row 17 has NAN value as it is reflecting true value in the Salinity column, 
#so we have to replace it either with 0 or do a interpolation
sel1.isnull() #this give the same result
```

```python
#command to interpolate the NAN values
#zero is not used to replace the NAN values because it will create a problem while ploting data 
sel1.Salinity=sel1.Salinity.interpolate()
```

```python
import matplotlib.pyplot as plt # to import the matplotlib library
# Temperature_C and Depth_m columns were selected
plt.scatter(sel1['Temperature_C'],sel1['Depth_metre'],c='#00FFFF',marker='*')# here '#00FFFF' is the cyan colour code  
plt.xlabel('Temperature in degree Celcius') #X axis
plt.ylabel('Depth in metre')# Y axis
plt.title('Temperature vs Depth')# title of the plot
plt.show()
```

```python
from sklearn.linear_model import LinearRegression #to import the linear regression library
```

```python
model=LinearRegression()# model is treated as object/variable
```

```python
#to select the x & y  values for the model
x=sel1[['Temperature_C']]
x.head(2)
```

```python
y=sel1['Depth_metre']
print(y.head(2))
print(type(y))#to print the return type of the object
```

```python
model.fit(x,y)# to fit the model using the input values x and y
```

```python
plt.scatter(x,y,c = '#7C5A5A') # command to display the scatter plot
plt.plot(x,model.predict(x))
plt.xlabel("Depth in metres")
plt.ylabel("Temperature in degree celcius")
plt.title("Variation of Tempterature with Depth" )
plt.show()
```

```python
model.score(x,y)*100 # command to check the model accuracy of predicting the values 
```
